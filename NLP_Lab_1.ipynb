Take your own text or take text as “The bank can guarantee deposits will
eventually cover future tuition costs because it invests in adjustable-rate
mortgage securities." Implement Ambiguity Removal in the text.

```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
def disambiguate_words(sentence):
    words = word_tokenize(sentence)
    disambiguated_words = []
    for word in words:
        synset = lesk(sentence, word)
        if synset:
            disambiguated_words.append(synset.name().split('.')[0])
        else:
            disambiguated_words.append(word)
    return ' '.join(disambiguated_words)
sentence = "The bank can guarantee deposits will eventually cover future tuition costs because it invests in adjustable-rate mortgage securities."
disambiguated_sentence = disambiguate_words(sentence)
print("Disambiguated Sentence:", disambiguated_sentence)
```

Take your own text or take text as “Hello there! How are you doing today?
NLP is fascinating." Implement Sentence Segmentation in the text.

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
def segment_sentences(text):
    return sent_tokenize(text)
def segment_words(sentence):
    return word_tokenize(sentence)
text = "Stocks are increasing in value. Traders are optimistic about market trends."
sentences = segment_sentences(text)
print("Sentences:", sentences)
for sentence in sentences:
    words = segment_words(sentence)
    print("Words in Sentence:", words)
```
