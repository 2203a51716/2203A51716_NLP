Implement Word Embeddings on following text

```python
import spacy
import gensim.downloader as api
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
words=['king','queen','man','woman','paris','france','london','england']
sentences=[words]
model=Word2Vec(sentences,vector_size=100,window=5,min_count=1,workers=4)
embeddings={word:model.wv[word]for word in words}
for word,vector in embeddings.items():
  print(f"Word:{word}\nEmbedding:{vector}\n")
```

Implement Word Embeddings using PCA (Principal Component Analysis) on following text

```python
words=['king','queen','man','woman','paris','france','london','england']
sentences=[words]
model=Word2Vec(sentences,vector_size=100,window=5,min_count=1,workers=4)
embeddings=[model.wv[word]for word in words]
pca=PCA(n_components=2)
result=pca.fit_transform(embeddings)
plt.figure(figsize=(8,6))
plt.scatter(result[:,0],result[:,1],c='green')
for i,word in enumerate(words):
  plt.annotate(word,xy=(result[i,0],result[i,1]))
plt.title("word embeddings visualized using PCA")
plt.show()
```

Execute Dependency Parsing on any text or on following text

```python
import spacy
nlp = spacy.load("en_core_web_sm")
sentence = "The quick brown fox jumps over the lazy dog."
doc = nlp(sentence)
for i,token in enumerate(doc):
    print(f"Word: {token.text}, POS: {token.pos_}, Dependency: {token.dep_}, Head: {token.head.text}")

```
